```python

def in_minutes(time):
    h, m = time.split(':')
    return int(h)*60+int(m)


def by_time(string):
    return string.split(' ')[3]


enter, leave = {}, {}
trips, trips_time = [], []
anser = 0

with open('input.txt') as file:
    log = [line.strip() for line in file]
    log = sorted(log, key=by_time)
    
for line in log:

    pid = line.split()[0]
    time_in_minutes = in_minutes(line.split()[3])

    if line.split()[1] == 'in' :
        enter[pid] = time_in_minutes
        
    else:
        leave[pid] = time_in_minutes
        trips.append((enter[pid],leave[pid]))

    # Один и тот же pid может совершить несколько поездок в течении дня. За счет сортировки вторая поездка 
    # сначала перепишет enter, а потом попадет в leave и даст новый элемент trips


trips_delta = [trip[1] - trip[0] for trip in trips]
anser = sum(trips_delta)/1110 #Количество минут между 5:00 и 23:30


with open("output.txt", "w") as out_file:
    out_file.write(str(anser))
```
